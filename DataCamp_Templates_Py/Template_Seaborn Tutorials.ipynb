### Step 1 Data

```python
import pandas as pd
import numpy as np
uniform_data=np.random.rand(10, 12)
print (uniform_data)
```

```python
data=pd.DataFrame({'x': np.arange(1, 101), 'y': np.random.normal(0, 4, 100)})
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Seaborn build-in dataset
titanic=sns.load_dataset("titanic")
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

### Step 3, plotting with seaborn

```python
# axis grids
g=sns.FacetGrid(titanic, col="survived", row="sex")
g=g.map(plt.hist, "age")
```

```python
sns.factorplot(x='pclass', y='survived', hue='sex', data=titanic)
```

```python
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=iris)
```

```python
h=sns.PairGrid(iris)
h=h.map(plt.scatter)
```

```python
sns.pairplot(iris)
```

```python
i=sns.JointGrid(x='x', y='y', data=data)
i=i.plot(sns.regplot, sns.distplot)
```

```python
sns.jointplot('sepal_length', 'sepal_width', data=iris, kind='kde')
```

```python
# Categorical plots
```

```python
# scatter plot
sns.stripplot(x='species', y='petal_length', data=iris)
```

```python
sns.swarmplot(x='species', y='petal_length', data=iris)
```

```python
# Bar chart
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
```

```python
# count plot
sns.countplot(x='deck', data=titanic, palette='Greens_d')
```

```python
# boxplot
sns.boxplot(x='alive', y='age', hue='adult_male', data=titanic)
```

```python
sns.boxplot(data=iris, orient='h')
```

```python
# violinplot
sns.violinplot(x='age', y='sex', hue='survived', data=titanic)
```

```python
# regression plot
sns.regplot(x='sepal_width', y='sepal_length', data=iris) # ax=ax, removed
```

```python
# distribution plot
plot=sns.distplot(data.y, kde=False, color='b')
```

```python
# maxtrix plot
sns.heatmap(uniform_data, vmin=0, vmax=1)
```

### Step 4 Customizations

```python
# axisgrid objects
g.despine(left=True)
g.set_ylabels('Survived')
g.set_xticklabels(rotation=45)
g.set_axis_labels('Survived', 'sex')

h.set(xlim=(0,5), ylim=(0,5), xticks=[0, 2.5, 5], yticks=[0, 2.5, 5])
```

```python
# plot
plt.title('A title')
plt.ylabel('Survived')
plt.xlabel('sex')
plt.ylim(0, 100)
plt.xlim(0, 10)
plt.setp(ax, yticks=[0,5])
plt.tight_layout()
# think more about this...
```
