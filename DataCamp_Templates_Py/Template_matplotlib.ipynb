## Matplotlib

### Step 1 Prepare the data

```python
# 1 D data
import numpy as np
```

```python
x=np.linspace(0, 10, 100)
y=np.cos(x)
z=np.sin(x)
```

```python
# 2D data or images
```

```python
data=2*np.random.random((10, 10))
data2=3*np.random.random((10,10))
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U=-1 - X**2 +Y
V= 1 + X - Y**2
```

```python
from matplotlib.cbook import get_sample_data
import matplotlib.image as mpimg
image = mpimg.imread('IMG_7251.png')
```

### Step 2 create plot

```python
import matplotlib.pyplot as plt
```

```python
# figure
fig=plt.figure()
fig2=plt.figure(figsize=plt.figaspect(2.0))

# axes
fig.add_axes()
ax1=fig.add_subplot(221) # row, col, number
ax3=fig.add_subplot(212)
fig3, axes=plt.subplots(nrows=2, ncols=2)
fig4, axes2 = plt.subplots(ncols=3)

# 1 D data
fig, ax = plt.subplots()
lines=ax.plot(x, y)
ax.scatter(x, y)
axes[0,0].bar([1,2,3], [3,4,5])
axes[1,0].barh([0.5,1,2.5],[0,1,2])
axes[1,1].axhline(0.45)
axes[0,1].axvline(0.65)
ax.fill(x, y, color='blue')
ax.fill_between(x, y, color='yellow')

# vector fields
axes[0,1].arrow(0,0,0.5, 0.5)
axes[1,1].quiver(y,z)
axes[0,1].streamplot(X, Y, U, V)

# data distribution
ax1.hist(y)
ax3.boxplot(y)
ax3.violinplot(z)

# 2D image
# img=np.load(get_sample_data('axes_grid/bivariate_normal.npy'))
im=ax.imshow(image, cmap='gist_earth', interpolation='nearest', vmin=-2, vmax=2) # got a local file

axes2[0].pcolor(data2)
axes2[0].pcolormesh(data)
CS=plt.contour(Y, X, U)
axes2[2].contourf(data2)
axes2[2]=ax.clabel(CS)
```

```python
# data
x=[1,2,3,4]
y=[10, 20, 25, 30]
# configure layout
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter([2,4,6], [5,15,25], color='darkgreen', marker='^')
ax.set_xlim(1, 6.5)
plt.show()
```

```python
### more from seaborn
import seaborn as sns
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
plt.hist(iris['sepal_length'], bins=30)
plt.show()
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.title('scatter plot')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.xticks([4, 5, 6, 7, 8 ], rotation=45)
plt.show()
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', hue='species', data=iris, fit_reg=False, legend=False)
plt.legend()
plt.show()
```

```python
# another try on figures and axes
fig, ax = plt.subplots()
ax.scatter(iris['sepal_length'], iris['sepal_width'])
plt.show()
```
